#### Objective : From the given iris dataset ,predict the optimal number of clusters and represent it visually.

```python
 #Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets

# Load the iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head() 
```

```python
# Creating array.
x = df.iloc[ : ,[0,1,2,3]].values
```

```python
np.array(x)
```

```python
# Importing liberary.
from sklearn.cluster import KMeans
sse=[] # Creating  array of squares sum of error
k_rng = range(1,11)
for k in k_rng :
    kmeans =KMeans(n_clusters = k)
    kmeans.fit(x)
    sse.append(kmeans.inertia_)
```

```python
# Plottig the line plot for 'The elbow method' to identify the optimal no. of clusters (k).
plt.figure(figsize=(10,5))
plt.plot(k_rng,sse)
plt.title('The elbow method')
plt.xlabel('No. of clusters(k)')
plt.ylabel('sum of squares error(sse)')
plt.grid()
plt.show
```

##### From the above line graph of 'The elbow method',we can conclude that sse is decreasing significantly onward k=3.
##### So , the optimal no. of clusters for this distribution is k= 3.

```python
# Predicting the value for no. of cluster (k)=3.
kmeans = KMeans(n_clusters=3)
y_predict = kmeans.fit_predict(x)
```

```python
np.array(y_predict)
```

```python
#Identify the centroid .
kmeans.cluster_centers_
```

```python
df['Cluster']=y_predict
df.head()
```

```python
df1 = df[df.Cluster==0]
df2 = df[df.Cluster==1]
df3 = df[df.Cluster==2]
```

```python
df1.head()
```

```python
#Plotting the scatter plot for the distribution of iris flowers and centroid.
# Here I am taking first two rows :sepal length & sepal width.
plt.figure(figsize=(10,5))
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='green',s=100,label='Iris-setosa')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='blue',s=100,label='Iris-versicolour')
plt.scatter(df3['sepal length (cm)'],df3['sepal width (cm)'],color='yellow',s=100,label='Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black',s=100,marker='*',label='Centoroids')
plt.legend()
```
